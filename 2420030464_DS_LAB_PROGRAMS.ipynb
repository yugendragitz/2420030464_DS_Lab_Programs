```python
#edit based measures hamming distance
def hamming_distance(str1, str2):

    if len(str1) != len(str2):
        raise ValueError("Strings must be of equal length")


    return sum(ch1 != ch2 for ch1, ch2 in zip(str1, str2))


s1 = "karolin"
s2 = "kathrin"

dist = hamming_distance(s1, s2)
print(f"Hamming Distance between '{s1}' and '{s2}': {dist}")
```

```python
#jaccard index
def jaccard_index(str1,str2):
  set1,set2 = set(str1.split()),set(str2.split())
  intersection = set1.intersection(set2)
  union = set1.union(set2)
  return len(intersection)/len(union)
s1 = "data science is fun"
s2 = "science makes data useful"
print("Jaccard Index:",jaccard_index(s1,s2))
```

```python
#longest common subsequence
def lcs_length(X, Y):
  m, n = len(X), len(Y)
  dp = [[0] * (n + 1) for _ in range(m + 1)]
  for i in range(m):
    for j in range(n):
      if X[i] == Y[j]:
        dp[i + 1][j + 1] = dp[i][j] + 1
      else:
        dp[i + 1][j + 1] = max(dp[i][j + 1], dp[i + 1][j])
  return dp[m][n]

seq1 = "ABCDEF"
seq2 = "AEBDF"

length = lcs_length(seq1, seq2)

print("Longest Common Subsequence Length:", length)

```

```python
# LU Decomposition using Doolittle Method
import numpy as np
def lu_decomposition(A):
    n = len(A)
    # Initialize L and U matrices
    L = np.zeros((n, n))
    U = np.zeros((n, n))
    for i in range(n):
        # Upper Triangular Matrix
        for k in range(i, n):
            sum_upper = 0
            for j in range(i):
                sum_upper += L[i][j] * U[j][k]
            U[i][k] = A[i][k] - sum_upper
        # Lower Triangular Matrix
        L[i][i] = 1  # Diagonal as 1
        for k in range(i + 1, n):
            sum_lower = 0
            for j in range(i):
                sum_lower += L[k][j] * U[j][i]
            L[k][i] = (A[k][i] - sum_lower) / U[i][i]
    return L, U
def forward_substitution(L, b):
    n = len(L)
    y = np.zeros(n)
    for i in range(n):
        sum_value = 0
        for j in range(i):
            sum_value += L[i][j] * y[j]
        y[i] = b[i] - sum_value
    return y
def backward_substitution(U, y):
    n = len(U)
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        sum_value = 0
        for j in range(i + 1, n):
            sum_value += U[i][j] * x[j]
        x[i] = (y[i] - sum_value) / U[i][i]
    return x
# Input Matrix A
A = np.array([
    [2, -1, -2],
    [-4, 6, 3],
    [-4, -2, 8]
], dtype=float)

b = np.array([-2, 9, -5], dtype=float)

L, U = lu_decomposition(A)

y = forward_substitution(L, b)

x = backward_substitution(U, y)

print("Matrix A:")
print(A)
print("\nLower Triangular Matrix L:")
print(L)
print("\nUpper Triangular Matrix U:")
print(U)
print("\nIntermediate Solution y:")
print(y)
print("\nFinal Solution x:")
print(x)
```

```python
# QR Decomposition using Gram-Schmidt Method
import numpy as np
def qr_decomposition(A):
    n = A.shape[0]

    Q = np.zeros((n, n))
    R = np.zeros((n, n))
    for j in range(n):
        v = A[:, j]
        for i in range(j):
            R[i, j] = np.dot(Q[:, i], A[:, j])
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        Q[:, j] = v / R[j, j]
    return Q, R
def solve_qr(Q, R, b):

    y = np.dot(Q.T, b)

    n = R.shape[0]
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        sum_value = 0
        for j in range(i + 1, n):
            sum_value += R[i, j] * x[j]

        x[i] = (y[i] - sum_value) / R[i, i]
    return x

A = np.array([
    [1, 1, 0],
    [1, 0, 1],
    [0, 1, 1]
], dtype=float)

b = np.array([4, 5, 6], dtype=float)

Q, R = qr_decomposition(A)

x = solve_qr(Q, R, b)

print("Matrix A:")
print(A)
print("\nOrthogonal Matrix Q:")
print(Q)
print("\nUpper Triangular Matrix R:")
print(R)
print("\nSolution Vector x:")
print(x)

```

```python
#Common Discrete Probability Distributions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson, nbinom, randint

# BERNOULLI DISTRIBUTION
p = 0.6
x = np.array([0, 1])
pmf = bernoulli.pmf(x, p)
print("Possible Values:", x)
print("PMF:", pmf)
samples = bernoulli.rvs(p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(5,4))
plt.bar(x, pmf)
plt.title("Bernoulli Distribution")
plt.xlabel("X")
plt.ylabel("Probability")
plt.show()
# BINOMIAL DISTRIBUTION
n = 10
p = 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
print("PMF:")
print(pmf)
samples = binom.rvs(n, p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
# GEOMETRIC DISTRIBUTION
p = 0.3
x = np.arange(1,11)
pmf = geom.pmf(x, p)
print("PMF:")
print(pmf)
samples = geom.rvs(p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Geometric Distribution")
plt.xlabel("Trial")
plt.ylabel("Probability")
plt.show()
# POISSON DISTRIBUTION
lam = 4
x = np.arange(0,15)
pmf = poisson.pmf(x, lam)
print("PMF:")
print(pmf)
samples = poisson.rvs(lam, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Poisson Distribution")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.show()
# NEGATIVE BINOMIAL DISTRIBUTION
r = 5
p = 0.4
x = np.arange(0,15)
pmf = nbinom.pmf(x, r, p)
print("PMF:")
print(pmf)
samples = nbinom.rvs(r, p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Negative Binomial Distribution")
plt.xlabel("Failures")
plt.ylabel("Probability")
plt.show()
# DISCRETE UNIFORM DISTRIBUTION
low = 1
high = 7
x = np.arange(low, high)
pmf = randint.pmf(x, low, high)
print("PMF:")
print(pmf)
samples = randint.rvs(low, high, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(6,4))
plt.bar(x, pmf)
plt.title("Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
#Common Continuous Probability Distributions
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import (
    norm,
    uniform,
    expon,
    gamma,
    beta,
    lognorm,
    chi2,
    t,
    weibull_min
)
# 1. NORMAL DISTRIBUTION
mu = 50
sigma = 10
x = np.linspace(20, 80, 200)
pdf = norm.pdf(x, mu, sigma)
samples = norm.rvs(mu, sigma, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 2. UNIFORM DISTRIBUTION
a = 0
b = 10
x = np.linspace(a, b, 200)
pdf = uniform.pdf(x, a, b)
samples = uniform.rvs(a, b, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Uniform Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 3. EXPONENTIAL DISTRIBUTION
print("\n========== Exponential Distribution ==========")
scale = 2
x = np.linspace(0, 15, 200)
pdf = expon.pdf(x, scale=scale)
samples = expon.rvs(scale=scale, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Exponential Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 4. GAMMA DISTRIBUTION
shape = 2
x = np.linspace(0, 15, 200)
pdf = gamma.pdf(x, a=shape)
samples = gamma.rvs(a=shape, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Gamma Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 5. BETA DISTRIBUTION
print("\n========== Beta Distribution ==========")
a = 2
b = 5
x = np.linspace(0,1,200)
pdf = beta.pdf(x, a, b)
samples = beta.rvs(a, b, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Beta Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 6. LOGNORMAL DISTRIBUTION
sigma = 0.5
x = np.linspace(0.01,5,200)
pdf = lognorm.pdf(x, sigma)
samples = lognorm.rvs(sigma, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Log-Normal Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 7. CHI-SQUARE DISTRIBUTION
df = 5
x = np.linspace(0,20,200)
pdf = chi2.pdf(x, df)
samples = chi2.rvs(df, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Chi-Square Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 8. STUDENT'S T DISTRIBUTION
df = 10
x = np.linspace(-5,5,200)
pdf = t.pdf(x, df)
samples = t.rvs(df, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Student's t Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 9. WEIBULL DISTRIBUTION
shape = 2
x = np.linspace(0,5,200)
pdf = weibull_min.pdf(x, shape)
samples = weibull_min.rvs(shape, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Weibull Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# Statistical Hypothesis Testing

# Application: Student Performance Analysis
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ztest

alpha = 0.05

# Z-Test
print("----- Z-Test -----")

sample_marks = np.array([78, 82, 85, 88, 90, 76, 95, 89, 84, 91])
population_mean = 80

z_stat, z_p_value = ztest(sample_marks, value=population_mean)

print("Z-Statistic:", z_stat)
print("P-Value:", z_p_value)

if z_p_value < alpha:
    print("Result: Reject Null Hypothesis")
else:
    print("Result: Fail to Reject Null Hypothesis")

# Independent t-Test
print("\n----- Independent t-Test -----")

class_A = np.array([78, 82, 85, 88, 90])
class_B = np.array([72, 75, 70, 68, 74])

t_stat, t_p_value = stats.ttest_ind(class_A, class_B, equal_var=False)

print("t-Statistic:", t_stat)
print("P-Value:", t_p_value)

if t_p_value < alpha:
    print("Result: Reject Null Hypothesis")
else:
    print("Result: Fail to Reject Null Hypothesis")
```

```python
#Parametric Statistical Tests
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.proportion import proportions_ztest

# Sample Data
np.random.seed(42)
group1 = np.random.normal(loc=70, scale=10, size=30)
group2 = np.random.normal(loc=75, scale=10, size=30)
group3 = np.random.normal(loc=80, scale=10, size=30)
before = np.random.normal(loc=65, scale=8, size=30)
after = before + np.random.normal(loc=5, scale=4, size=30)

# 1. ONE-SAMPLE Z-TEST
z_stat, p_value = ztest(group1, value=70)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
# 2. TWO-SAMPLE Z-TEST
z_stat, p_value = ztest(group1, group2)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
# 3. ONE-PROPORTION Z-TEST
success_count = 60
sample_size = 100
claimed_proportion = 0.50
z_stat, p_value = proportions_ztest(
    count=success_count,
    nobs=sample_size,
    value=claimed_proportion
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
# 4. TWO-PROPORTION Z-TEST
successes = np.array([60, 75])
samples = np.array([100, 120])
z_stat, p_value = proportions_ztest(
    count=successes,
    nobs=samples
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
# 5. ONE-SAMPLE T-TEST
t_stat, p_value = stats.ttest_1samp(group1, popmean=70)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# 6. INDEPENDENT TWO-SAMPLE T-TEST
t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# 7. PAIRED SAMPLE T-TEST
t_stat, p_value = stats.ttest_rel(before, after)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# 8. ONE-WAY ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)
print("F-statistic:", f_stat)
print("P-value:", p_value)
# 9. PEARSON CORRELATION TEST
x = np.random.normal(50, 10, 30)
y = x + np.random.normal(5, 5, 30)
r_value, p_value = stats.pearsonr(x, y)
print("Correlation Coefficient:", r_value)
print("P-value:", p_value)
# 10. SIMPLE LINEAR REGRESSION TEST
slope, intercept, r_value, p_value, std_error = stats.linregress(x, y)

print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r_value)
print("P-value:", p_value)
print("Standard Error:", std_error)
# 11. F-TEST FOR VARIANCE
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)
f_stat = var1 / var2
df1 = len(group1) - 1
df2 = len(group2) - 1
p_value = 2 * min(
    stats.f.cdf(f_stat, df1, df2),
    1 - stats.f.cdf(f_stat, df1, df2)
)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
import numpy as np
from scipy import stats
# Sample Data
np.random.seed(42)

group1 = np.array([45, 50, 52, 55, 58, 60, 62, 65])
group2 = np.array([48, 53, 57, 59, 61, 64, 67, 70])
group3 = np.array([40, 44, 49, 51, 54, 56, 59, 63])

before = np.array([50, 52, 55, 57, 60, 62, 64, 66])
after  = np.array([53, 55, 56, 60, 63, 65, 67, 70])

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 4, 5, 7, 8, 10, 11, 13])

observed = np.array([25, 30, 45])
expected = np.array([33.33, 33.33, 33.34])

# 1. MANN-WHITNEY U TEST

u_stat, p_value = stats.mannwhitneyu(group1, group2)

print("U-statistic:", u_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")

# 2. WILCOXON SIGNED-RANK TEST

w_stat, p_value = stats.wilcoxon(before, after)

print("W-statistic:", w_stat)
print("P-value:", p_value)

# 3. KRUSKAL-WALLIS H TEST

print("\n========== Kruskal-Wallis H Test ==========")

h_stat, p_value = stats.kruskal(group1, group2, group3)

print("H-statistic:", h_stat)
print("P-value:", p_value)

# 4. FRIEDMAN TEST

print("\n========== Friedman Test ==========")

condition1 = np.array([55, 60, 62, 64, 66])
condition2 = np.array([58, 61, 65, 67, 69])
condition3 = np.array([60, 63, 66, 70, 72])

friedman_stat, p_value = stats.friedmanchisquare(
    condition1, condition2, condition3
)

print("Friedman Statistic:", friedman_stat)
print("P-value:", p_value)


# 5. SPEARMAN RANK CORRELATION

rho, p_value = stats.spearmanr(x, y)

print("Spearman Correlation:", rho)
print("P-value:", p_value)

# 6. KENDALL TAU CORRELATION

tau, p_value = stats.kendalltau(x, y)

print("Kendall Tau:", tau)
print("P-value:", p_value)

# 7. CHI-SQUARE GOODNESS OF FIT TEST

print("\n========== Chi-Square Goodness of Fit Test ==========")

chi_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)

# 8. CHI-SQUARE TEST OF INDEPENDENCE

table = np.array([
    [30, 10],
    [20, 40]
])

chi_stat, p_value, dof, expected_freq = stats.chi2_contingency(table)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected_freq)

# 9. KOLMOGOROV-SMIRNOV TEST

sample = np.random.uniform(0, 1, 100)

ks_stat, p_value = stats.kstest(sample, 'uniform')

print("KS Statistic:", ks_stat)
print("P-value:", p_value)

# 10. MEDIAN TEST

stat, p_value, median, contingency_table = stats.median_test(
    group1, group2, group3
)

print("Statistic:", stat)
print("P-value:", p_value)
print("Grand Median:", median)
print("Contingency Table:")
print(contingency_table)

# 11. SIGN TEST

differences = after - before

positive_signs = np.sum(differences > 0)
negative_signs = np.sum(differences < 0)

n = positive_signs + negative_signs

p_value = stats.binomtest(
    positive_signs,
    n,
    p=0.5,
    alternative='two-sided'
).pvalue

print("Positive Signs:", positive_signs)
print("Negative Signs:", negative_signs)
print("P-value:", p_value)
```

```python
# Classification using Decision Tree

import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)


X = np.array([
    [2, 50],
    [3, 60],
    [4, 65],
    [5, 70],
    [6, 75],
    [7, 80],
    [8, 85],
    [9, 90]
])

y = np.array([0, 0, 0, 1, 1, 1, 1, 1])

model = DecisionTreeClassifier(random_state=42)

model.fit(X, y)

X_test = np.array([
    [3, 55],
    [5, 72],
    [8, 88],
    [4, 60]
])

y_test = np.array([0, 1, 1, 0])

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
error_rate = 1 - accuracy
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)


print("Actual Labels:")
print(y_test)
print("\nPredicted Labels:")
print(y_pred)
print("\nConfusion Matrix:")
print(cm)
print("\nAccuracy:")
print(accuracy)
print("\nError Rate:")
print(error_rate)
print("\nPrecision:")
print(precision)
print("\nRecall:")
print(recall)
print("\nF1-Score:")
print(f1)
```

```python
# Text Analytics and Recommendation System


import pandas as pd
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')


print("----- Text Preprocessing -----")

text = "The movie was amazing, exciting, and full of action scenes!"

text = text.lower()

text = text.translate(str.maketrans('', '', string.punctuation))

tokens = word_tokenize(text)

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

print("Original Text:")
print("The movie was amazing, exciting, and full of action scenes!")

print("\nProcessed Tokens:")
print(filtered_tokens)


print("\n----- Content-Based Recommendation -----")


movies = pd.DataFrame({
    'Movie': [
        'Avengers',
        'Batman',
        'Superman',
        'Titanic',
        'Inception'
    ],
    'Description': [
        'action superhero marvel fight',
        'dark superhero action detective',
        'superhero flying action hero',
        'romance ship love drama',
        'dream science fiction thriller'
    ]
})

print("\nMovie Dataset:")
print(movies)

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(movies['Description'])

similarity = cosine_similarity(tfidf_matrix)

movie_name = "Batman"

if movie_name in movies['Movie'].values:

    movie_index = movies[movies['Movie'] == movie_name].index[0]


    scores = list(enumerate(similarity[movie_index]))

    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nRecommended Movies for {movie_name}:")

    recommendations = 0


    for i in sorted_scores[1:]:
        if i[1] > 0:
            print(f"{movies.iloc[i[0]]['Movie']} (Similarity: {i[1]:.2f})")
            recommendations += 1

        if recommendations == 3:
            break

    if recommendations == 0:
        print("No similar movies found.")

else:
    print("Movie not found in the dataset.")
```
